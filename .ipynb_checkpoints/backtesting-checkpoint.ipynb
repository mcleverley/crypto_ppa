- if bitcoin_price is 1000, and you buy \$100 USD worth of BTC
    - you get 1/10th of a bitcoin. how do we math this?
    - purchased_btc = 100 / 1000 = 1/10 = spent_USD / BTC_price
- if btc_price = 1000, and you have one bitcoin & sell half of it, you get $500 usd
    - purchased_usd = spent_BTC * BTC_price
    - 500 = 0.5(BTC) * 1000

```python
class portfolio:
    # store a current_balance of BTC and USD + current bitcoin_price (relative USD)
    stops = [] # empty list of auto-stops
    def __init__(self, usd, btc, price, duration):
        self.usd = usd # $dollars in portfolio
        self.btc = btc # number of bitcoins in portfolio
        self.price = price # current price of bitcoin/usd
#         self.duration = duration # how long to run the portfolio for
    
    # buy/sell equations. the mirrored n/ n* is sort of elegant, didn't plan that one
    def buy(self, n): # convert $n USD to BTC
        self.usd -= n # spend $n
        self.btc += (n/self.price) # to purchase BTC
        
    def sell(self, n): # convert n BTC to USD (both are stored in their own units)
        self.usd += n*self.price # sell n bitcoins at price = usd_gained
        self.btc -= (n) # by spending BTC
        
class auto_stop:
    # stores auto buy or sell point that a class can possess multiple of (or none)
    def __init__(self, initial, direction, thresh, period):
        self.initial = initial # BTC price at which stop was set against
        self.direction = direction # 
        self.thresh = thresh # threshold at which to buy/sell
        self.period = period # number of rows remaining in active period of PPA



p = portfolio(2000, 1, 1000) # $2000 USD starting, no BTC. 1btc / $1000 USD
print(p.usd,p.btc)

p.buy(0.1*p.usd)
print(p.usd,p.btc)

p.sell(0.1*p.btc)
print(p.usd,p.btc)
```

```python
alerts = [col for col in list(df.columns) if 'alert' in col]
def backtest(p, df):
    cash = [] # storage of portfolio's USD & BTC over time
    coin = [] # index-aligned with the dataframe for easy graphing
    for i, row in tqdm_notebook(df.iterrows()): # loop through dataframe
        # iterative maintenance
        p.price = row['close'] # update BTC price with this row's data
        # check for any PPAs activating this row
        for alert in alerts:
            if row[alert] == 1:  # if PPA activates now
                ppa = True
                # set length, period, direction of alert
                length = alert[:alert.index('min')] # 5, 30 or 60
                if length == '5': period = 15
                else: period = int(length) * 15 # if i did the math right this should work
                if 'long' in alert: 
                    direction = 1
                    direct = 'long'
                else: 
                    direction = -1
                    direct = 'short'
                # get conf
                conf_dict = {}
#                 '5min_short_10pct_conf' # target
#                 for i
#                 tar = str(length)+'min_'+direct
                for col in list(df.columns):
                    if (str(length)+'min_'+direct in col and 'pct' in col):
#                     if str(length) in col and 'pct' in col and direct in col: # one of four %s for that length/direction
#                         print('at index ' + str(i))
#                         print(col)
#                         print(alert, row[alert])
#                         print(row['timestamp'])
                        ind = col.index('pct')
                        conf = col[ind-2:ind] # 10, 25, 50 or 75
                        try:
                            val = int(row[col])
                            conf_dict[conf] = val # assign (conf: val) pair to dict
                        except Exception as e:
#                             print(e)
                            ppa = False
#                             print('')
                        
                # confidence values recorded
                # if PPA: buy/sell according to direction
                if ppa:
#                     print('buying/selling based on PPA direction')
                    investment = 0.1 # ratio. invest one-tenth of your wealth, predicting the spike
                    initial = row['close'] # initial BTC_price at which stop is set
#                     print(conf_dict)
                    if direct=='long':
                        p.buy(p.usd * investment)
                        p.stops.append(auto_stop(initial, direction, conf_dict["75"], period))
                    elif direct=='short':
                        p.sell(p.btc * investment)
                        p.stops.append(auto_stop(initial, direction, conf_dict["75"], period))
            # end PPA activity
        # regardless of PPA active or not:
        # if stops active: de-increment each
            # check if current BTC_price has passed any thresholds of stops; if so, sell/buy accordingly
        if p.stops: # if active stops, de-increment its period by one (row), simulate passing of time
            for stop in p.stops:
                stop.period -= 1
                if stop.period <= 0:
                    p.stops.remove(stop) # delete expired stops
                # check if price has eclipsed the threshold of any stop
                if (stop.direction=='long' and row['close'] >= stop.initial+stop.thresh):
                    # your auto-stop long gambit paid off, you lucker dog. time to sell
                    p.sell(p.btc * investment)
                elif (stop.direction=='short' and row['close'] <= stop.initial-stop.thresh):
                    # your short strategy actually came through, time to buy
                    p.buy(p.usd * investment)
        cash.append(p.usd)
        coin.append(p.btc)
    # end function
    return cash, coin

# create new portfolio initial monies
p = portfolio(10000, 5, df.iloc[189228]['close']) # change the iloc according to start point


old_df = df
# df = df[189228:] # start halfway through the period (well after the initial $1k -> 20k spike)
cash, coin = backtest(p, df)
                
```

```python
def draw_portfolio(df): # dataframe. subset the timestamps to zoom in on an area

    sns.set(style="darkgrid", rc={'figure.figsize':(35,19)})
    sns.set_context('poster')
    
    w = 1 # line width
    # tinkering with alpha levels to make it more readable
    ax = df.plot(x="timestamp", y='cash', color='forestgreen', alpha=0.7, legend=False,linewidth=w)

    ax2 = ax.twinx() # twin axes for multilines
    df.plot(x="timestamp", y='adj_coin', ax=ax2, color='darkorange', alpha=0.7, legend=False,linewidth=w)

    ax3 = ax.twinx()
    df.plot(x='timestamp', y='close', ax=ax3, color='indigo', alpha=0.4, legend=False,linewidth=w, 
#             tick_params={
#     axis:'y',          # changes apply to the x-axis
#     which:'both',      # both major and minor ticks are affected
#     bottom:False,      # ticks along the bottom edge are off
#     top:False,         # ticks along the top edge are off
#     labelbottom:False}
           )

    ax.figure.legend(loc='lower right', bbox_to_anchor=(0.15, 0.58, 0,0))
    plt.title('3.5 year Portfolio in USD (left) and Bitcoin (right) vs. BTC price',
              fontdict={'verticalalignment': 'center',})
    # horizontal lines denote starting investment
    ax.axhline(y=df.iloc[0]['cash'], color='forestgreen')
    ax2.axhline(y=df.iloc[0]['adj_coin'], color='darkorange')

    plt.show()
    return

last = sub[-1000:]
draw_portfolio(sub)
```

```python
def calc_gross(row): # gross wealth of 'coin' and 'cash' in usd
    # just multiply 'coin' by 'closing_price' and add to cash
    gross = row['cash'] + (row['coin']*row['close'])
    return gross
df['gross'] = df.apply(lambda row: calc_gross(row), axis=1)
```

## lingering concerns
- the question isn't "how well does it do", but "how well does it do compared to what's happening in the market"
- a 450% increase from Jan 2017 to Aug 2020 is good, but considering the price of bitcoin spiked from \$1k to \$20k in 2018, it's not amazing
    - consider that wealth only grew from \$42k to \$52k if we start with the same USD and num_bitcoins in Nov 2018 (starting at second half of the period)
    - ways to combat this:
        - start with 0.1 bitcoins at beginning to lessen the "spike" growth
        - dynamic 'investment' numbers
            - investment = base + intensity
                - intensity calculated by considering "how dramatic is the 75% prediction value compared to previous 75% values of this same PPA type in the past"
                    - sell more on a more drastic drop, etc
                - could also look at how much the 75, 50, 25, 10 are 'close' or 'far' from each other, or how dramatic the various thresholds are compared to normal
        - sequester off a portion of funds every time auto-stop buys or sells on the predicted threshold
            - 'stable' money that doesn't invest, but is protected against price changes
- 25% wealth growth in 1.5 years isn't bad, compared to index funds around 5-7%
- but this algorithm could be a lot better

# rolling x-month portfolios
- every day, begin a new portfolio within N USD & M BTC (or N worth of BTC)
- each portfolio lasts three months
- each portfolio follows the same logic (bets 10% of its current wealth on the PPA's 75% auto-stop)
we'll have like ~1,000 3-months to average the gains of. should account for volatility / overspecific conditions

```python


def rolling_backtest(df, days=90, usd=3000, btc=2, investment=0.1): # specify portfolio & investment pct
    # returns list of portfolio objects
    ports = [] # holds list of active portfolios
    finished_ports = [] # concluded portfolios
    alerts = [col for col in list(df.columns) if 'alert' in col] # list of price alerts
    columns = list(df.columns)
#     moves = {} # record when auto-stops pay off. {'date': when we moved, ''}
    # loop through each row
    for i, row in tqdm_notebook(df.iterrows()):
        # assign current row's BTC price
        current_price = row['close']
        
        # if new day, create new portfolio
        if t.hour == 0 and t.minute == 0: # ~ dawn of a new day ~ 72 hours remain ~
            # 1440 minutes per day, 5 minutes per row
            rows = (days * 1440) / 5 # other new_portfolio vars from above function params
            p = Portfolio(usd, btc, row['close'], duration=rows) # new portfolio with initial wealth, btc_price & duration
            ports.append(p) 
            
        # loop through each active portfolio
        for p in ports:
            # de-increment port lifetime
            p.duration -= 1
            if p.duration <= 0: # move expired portfolios to other list
                finished_ports.append(p)
                ports.remove(p) # is this 'cleaner' than keeping them in the same list
                # and only iterating portfolios with duration > 0? less checks with this method i think
            # check if any PPAs are activating
            for alert in alerts:
                if row[alert] == 1:
                    ppa = True
                    # set alert length, period, duration
                    length = alert[:alert.index('min')]
                    period = int(length) * 3 # period: num_rows PPA is active for
                                # worked this out after first run. 5-min PPA is active for 75 minutes or 15 rows
                                # 30-min PPA active for 7.5 hours, 450 mins, 90 rows
                                # 60-min PPA active for 15 hours, 900 mins, 180 rows
                                # the coefficient is 3, not 15 like my first tests.
                        # interesting that i still got good returns with grievously miscalculated auto-stop lengths
                        # will re-run those later
#                     if length == '5': period = 15 
                        # 30- minute PPA should be active for 7.5 hours = 450 minutes = 90 5-minute rows
                        # 60 - minute PPA should be active for 15 hours = 900 mins = 180 5-minute rows
                        # 90 / 30 = 3; 180 / 60 = 3; 15 = 5 = 3
                            # ah, i missed that in the first go-around, multiplied by 15 instead of 3
                            # could have just specified, but it's nice to do it programatically
#                     else: period = int(length) * 3 # if length=60, 60*15 = 900; 900/5
                    if 'long' in alert:
                        direction = 1
                        direct = 'long'
                    else:
                        direction = -1
                        direct = 'short'
                    # get confidence dict of alert
                    conf_dict = {}
                    for col in columns:
                        if (str(length)+'min_'+direct in col and 'pct' in col): # confidence value
                            ind = col.index('pct')
                            conf = col[ind-2:ind] # 10, 25, 50 or 75% confidence interval
                            try:
                                val = int(row[col]) # the predicted price change for said interval
                                conf_dict[conf] = val # assign pair to dict
                            except Exception as e:
                                print(e)
                                ppa = False # ppa activated, but didn't generate predictions (too early)
                    # if ppa generated predictions: bet on it
                    if ppa: 
                        initial = row['close'] # initial price to set stop
                        if direct == 'long':
                            p.buy(p.usd * investment)
                            p.stops.append(auto_stop(initial, direction, conf_dict['75'], period))
                        elif direct == 'short':
                            p.sell(p.btc * investment)
                            p.stops.append(auto_stop(initial, direction, conf_dict['75'], period))
                # regardless of PPA active or not
                if p.stops: # if portfolio has any active stops
                    for stop in p.stops:
                        stop.period -= 1 # de-increment active lifetime of stop
                        if stop.period <= 0:
                            p.stops.remove(stop) # delete expired stop)
                            continue # don't trade on expired stop
                        # if current_price passes stop threshold, trade on it you lucky dog
                        if (stop.direction=='long' and row['close'] >= stop.initial+stop.thresh):
                            # your long gambit paid off, sell some coin
                            p.sell(p.btc * investment) # record this win in your logbook
#                             # compare how much you just sold for vs. how much you invested on the stop
#                             stop_money = 
#                             gain = 
#                             move = {'date':row['date'], 'direction':stop.direction,
#                                    'price':row['close'], ''}
                        # i could just calculate this later from the movement of portfolios, probably
                            # save on runtime
                        elif (stop.direction=='short' and row['close'] >= stop.initial+stop.thresh):
                            # you shorted the market like a pro, buy some coin
                            p.buy(usd * investment)
            # append current wealth totals to this portfolio's tracker lists
            # (whether or not you moved any money this row)
            p.cash.append(p.usd)
            p.coin.append(p.btc)
            # end all_portfolios loop
        # end df loop
    return ports
        
            
                
                            
            
```
